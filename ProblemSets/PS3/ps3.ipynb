```python
import sympy as sy
w1 = 1.0
w2 = sy.symbols('w2')
w3 = sy.symbols('w3')
beta = 0.9
solution = sy.solve([sy.log(w2 - w3).diff(w3) + beta * sy.log(w3).diff(w3), sy.log(w1 - w2).diff(w2) + beta * sy.log(w2 - w3).diff(w2)], dict=True)
c1 = w1 - solution[0][w2]
c2 = solution[0][w2] - solution[0][w3]
dictionary_c = {'c1':c1, 'c2':c2, 'c3':solution[0][w3]} 
dictionary_w = {'w1':w1, 'w2':solution[0][w2], 'w3':solution[0][w3], 'w4':0}
print(dictionary_c)
print(dictionary_w)
```

```python
import matplotlib.pyplot as plt
T = [0, 1, 2, 3]
C = [dictionary_c['c1'], dictionary_c['c2'], dictionary_c['c3']]
W = [dictionary_w['w1'], dictionary_w['w2'], dictionary_w['w3'], dictionary_w['w4']]
plt.plot(T, W)
plt.title('How W Evolves over Three Periods')
plt.xlabel('T')
plt.ylabel('W')
```

```python
from scipy.stats import norm
M = 7
sigma = 0.5
mu = 4 * sigma
E = np.linspace(mu - 3 * sigma, mu + 3 * sigma, M)
E
```

```python
pdf = lambda x: norm(loc = mu, scale = sigma).pdf(x)
pdf = pdf(E)
pdf
```

```python
plt.plot(E, pdf)
plt.show()
```

```python
# q17
```

```python
c_mat = (np.tile(w_vec.reshape((N, 1)), (1, N)) - 
         np.tile(w_vec.reshape((1, N)), (N, 1)))
c_pos = c_mat > 0
c_mat[~c_pos] = 1e-7
u_mat = utility(c_mat)
c = np.array([u_mat * e for e in E])

v_init = np.zeros((N, M))
expected_v = v_init @ pdf.reshape((M, 1))
expected_v_mat = np.tile(expected_v.reshape((1, N)), (N, 1))
expected_v_mat[~c_pos] = -9e+4
expected_v_cube = np.array([expected_v_mat for e in range(M)])

vt = beta * expected_v_cube + c
v_new = np.zeros((N, M))
w_new = np.zeros((N, M))
for i in range(N):
    array = vt[:, i, :]
    v_new[i] = array.max(axis = 1)
    arr = np.argmax(array, axis = 1)
    w_new[i] = w_vec[arr]
```

```python
plt.plot(w_vec, np.average(w_new, axis=1))
plt.title("policy function")
plt.show()
```

```python
plt.plot(np.average(v_new,axis=1))
plt.title("value function")
plt.show()
```
